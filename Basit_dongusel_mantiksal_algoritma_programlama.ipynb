#### Basit Günlük Kar-Zarar Hesaplama Programı

```python
basa_bas = 1000
beklenen_ciro = 1200
satis_gunluk = int(input("Günlük satışınızı giriniz: "))

if satis_gunluk > basa_bas:
    if satis_gunluk > beklenen_ciro:
        print("Beklenenden %", (satis_gunluk / beklenen_ciro) * 100 - 100, 
              " daha fazla satış yaptınız.", 
              "\n", 
              "Günlük karınız ise :", 
              satis_gunluk - basa_bas, 
              "\n", 
              " kar oranınız: ", ((satis_gunluk / basa_bas) * 100)- 100,
             "Beklene ciro: ", beklenen_ciro, "\n",
             "Kar-zarar eşiği: ", basa_bas)
    elif satis_gunluk == beklenen_ciro:
        print("Beklenen kadar satış yaptınız.","\n","Karınız: ", satis_gunluk - basa_bas,"\n",
             "Beklene ciro: ", beklenen_ciro, "\n",
             "Kar-zarar eşiği: ", basa_bas)
    else:
        print("beklenenin altında kar yaptnız", "\n",
             "Günlük karınız: ", satis_gunluk - basa_bas, "\n",
            "Beklene ciro: ", beklenen_ciro, "\n",
             "Kar-zarar eşiği: ", basa_bas)
else:
    print((satis_gunluk / basa_bas) * 100 - 100, "%" ,"zarardasınız.",
         "Beklene ciro: ", beklenen_ciro, "\n",
             "Kar-zarar eşiği: ", basa_bas)
```

```python


bese_bolunenler = []
bolunmeyenler = []

for i in range(0,100,1):
    if (i*2+1) % 5 == 0:
        bese_bolunenler.append(i)
        print(i)
    else:
        bolunmeyenler.append(i)
print("1 ile 100 arasında 2 katının 1 fazlası 5'e tam bölünen sayıların adedi : ",len(bese_bolunenler), "\n", "bölünmeyenlerin adedi", len(bolunmeyenler))
```

```python
###Derste sorulan sorunun cevabı

### Adım 1: Başla
### Adım 2: TANIMLA bolunenler BOŞ
### Adım 3: TANIMLA bolunmeyenler BOŞ
### Adım 4: TANIMLA sayac 0
### Adım 5: TANIMLA while < 200
### Adım 6: Eğer (i*2+1) % 5 = 0; Ekle i bolunenler
### Adım 7: Eğer (i*2+1) % 5 != 0; Ekle i bolunmeyenler
### Adım 8: i+=1
### Adım 9: Eğer i < 200; Dön Adım 6
### Adım 10: Bitir
bolunenler = []
bolunmeyenler = []

i = 0

while i <= 200:
    if (i*2+1) % 5 == 0:
        bolunenler.append(i)
    else:
        bolunmeyenler.append(i)
    i+=1
print("5'e Bölünen Sayılar: ", bolunenler, "\n", "\n",
      "Bölünmeyen Sayılar: ", bolunmeyenler)
```
